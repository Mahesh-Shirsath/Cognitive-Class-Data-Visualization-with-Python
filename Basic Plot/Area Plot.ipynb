```
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
```

```
dataset.set_index('OdName', inplace=True)
```

```
dataset.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
dataset['Total'] = dataset.sum(axis=1)
```

```
dataset.head()
```

```
dataset.sort_values('Total', ascending=False, inplace=True)
```

```
dataset.head()
```

```
years = list(range(1980,2014))
df_top5 = dataset.head()
```

```
df_top5 = df_top5[years].transpose()
```

```
df_top5.tail()
```

```
df_top5.plot(kind='line')
plt.title('Immigartion trend of top 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

As we can see the line plot is not providing the correct and important infromation thus we use area plot in such cases

```
df_top5.plot(kind='area', figsize=(20, 10))
plt.title('Immigartion trend of top 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```
df_top5.plot(kind='area', 
             stacked=False,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```
