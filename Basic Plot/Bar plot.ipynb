```
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
```

```
dataset.set_index('OdName', inplace=True)
```

```
dataset.drop(['Type',	'Coverage',	'AREA', 'REG', 'DEV'], axis=1, inplace=True)
```

```
dataset['Total'] = dataset.sum(axis=1)
```

```
dataset.head()
```

```
df_india = dataset.loc['India', range(1980,2014)]
```

```
df_india.plot(kind='bar')
plt.title('Indian immigrants to cananda')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```
