```
dataset = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

```
dataset.head()
```

```
years = list(range(1980,2014))
```

```
dataset.set_index('OdName', inplace=True)
```

```
dataset.loc['Haiti', years].plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.text(2000, 6000, '2010 Earthquake') # see note below
plt.xlabel('Years')
```

```
data = dataset.loc[['India','China'], years]
data.plot(legend=False)
```
